#K-MEANS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.describe()
```

```python
#Normalisation
def minmax_fun(i):
  x=(i-i.min())/(i.max()-i.min())
  return x
```

```python
norm_df=minmax_fun(crime.iloc[:,1:])
norm_df
```

```python
scaler=StandardScaler()
scaled_crime = scaler.fit_transform(crime.iloc[:,1:])
scaled_crime
```

ELBOW CHART

```python
wcss =[]
for i in range(1,11):
  kmeans =KMeans(n_clusters=i,random_state=0)
  kmeans.fit(norm_df)
  wcss.append(kmeans.inertia_)

  print(wcss)

plt.plot(range(1,11),wcss)
plt.title('Elbow Chart')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#build cluster algorithm
model=KMeans(5,random_state=0)
model
```

```python
pred_val =model.fit_predict(norm_df)
pred_val
```

```python
crime['model']=pred_val
crime
```

```python
crime.groupby('model').agg(['mean'])
```

```python
crime['model'].value_counts()
```

```python
#vishualisation
sns.displot(crime)
```

```python
sns.pairplot(crime,hue='model')
```

```python
sns.stripplot(crime)
```
